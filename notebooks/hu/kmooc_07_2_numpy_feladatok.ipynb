## 1. Egyszerű tartó lehajlása (statika, folytonos tartó közelítése)

Egy kéttámaszú tartó hossza L = 6 m. A tartót egyenletesen megoszló teher terheli: q = 12 kN/m. A lehajlás közelítő képlete egyszerűsített alakban:

$w(x) = \frac{q \, x \, \left(L^{3} - 2L x^{2} + x^{3}\right)}{24 E I}$

ahol:
- $q$ = 12 kN/m
- $E$ = 2.1e11 Pa (acél),
- $I$ = 8.5e-6 $m^4$,
- $x$ a támasztól mért távolság.

Feladat:
* Hozz létre egy x tömböt 0 és L között 301 ponttal (np.linspace).
* Számítsd ki a w lehajlást minden x pontra vektorosan (ciklus nélkül).
* Határozd meg a maximális lehajlást és azt, hogy hol következik be (w.max(), np.argmax).
* Alakítsd át a lehajlás értékeket milliméterre (mm), és írd ki a maximumot mm-ben.

````python
import numpy as np

# paraméterek
L = 6.0
q = 12_000.0      # N/m
E = 2.1e11        # Pa
I = 8.5e-6        # m^4

# ronda nagy képlet, hogy ne kelljen begépelned...
w = (q * x * (L**3 - 2 * L * x**2 + x**3)) / (24 * E * I)
````
