## V√°ltoz√≥k

Az adatokkal n√©ha el√©g neh√©z dolgozni, ha mindig nek√ºnk kell ≈ëket defini√°lni. N√©ha nem is lehet. Ha p√©ld√°ul ha a felhaszn√°l√≥t√≥l √©rkezik egy sz√°m, nem is tudhatjuk, hogy az pontosan mi lesz.

Sok programoz√°si nyelvben ezt "v√°ltoz√≥kkal" oldj√°k meg, azaz deklar√°lnak egy bizonyos t√≠pus√∫ v√°ltoz√≥t, ami abb√≥l a tipusb√≥l b√°rmilyen √©rt√©ket felvehet.

A Pythonban igaz√°b√≥l nincsnek v√°ltoz√≥k. B√°r sokan v√°ltoz√≥knak nevezik ≈ëket, ezek ink√°bb cimk√©k. Pythonban az adatnak van t√≠pusa, a cimk√©nek nincs. A cimk√©t b√°rmire r√°ragasztjatjuk, majd √©s √°tragaszthatjuk egy m√°sik t√≠pus√∫ adatra. A "c√≠mk√©z√©st" az egyenl≈ës√©gjellel (=) v√©gezz√ºk. A v√°ltoz√≥ (cimke) neve b√°rmi lehet, amit a Python maga nem haszn√°l, viszont nem kezd≈ëdhet sz√°mmal, kis nagybet≈±t is tartalmazhat (√©s meg is k√ºl√∂nb√∂zteti).


````python
kakadu = 67.23 # a lebeg≈ëpontos sz√°mhoz (adathoz) hozz√°rendelt√ºk a kakadu cimk√©t
kakadu = "csip?" # a kakadu cimk√©t √°tragasztottuk erre a sz√∂veges adatra
Kakadu = 100 # ez egy m√°sik cimke, mert nagybet≈±s, a kis kakadu tov√°bbra is a sz√∂veg

print(kakadu) # √°tadjuk a ki√≠r√≥ f√ºggv√©nynek a felcimk√©zett adatot
print(Kakadu) # ez a nagy Kakadu, m√°sik cimke, m√°sik adat.
````

A modern python teljes√©rt√©k≈± unicode t√°mogat√°ssal rendelkezik, √≠gy nyugodtan adhatunk ak√°r ilyen v√°ltoz√≥neveket is:

# Python Adatt√≠pusok


Az adatt√≠pus l√©nyeg√©ben annyit jelent, hogy a mem√≥ri√°ban t√°rolt bitmint√°t hogyan √©rtelmezz√ºk.
