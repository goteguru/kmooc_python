## math – matematikai alapműveletek

A legtöbb mérnöki számításhoz nélkülözhetetlen.
Tartalmaz: trigonometria, gyök, logaritmusok, konstansok (π, e).

```python
import math

d = 12
kerulet = math.pi * d

a,b = 2.32, 0.21

math.pi          # π
math.e           # Euler-szám
math.sqrt(a)     # négyzetgyök
math.pow(a, b)   # hatványozás

math.floor(a)   # lefelé kerekít
math.ceil(a)    # felfelé
math.fabs(a)    # abszolút érték

rad = 0.21
math.sin(rad)
math.cos(rad)
math.tan(rad)

fokban = math.degrees(rad)   # radián → fok
math.radians(fokban)   # fok → radián
# lesz benne egy kis hiba...

```

## statistics – alapvető statisztika
Ha nem szeretnél beindítani valami igazi nagyágyút (pl. numpy)

```python
import statistics as st
data = [10,12,9,14]

statisztikák = {
  'átlag': st.mean(data),
  'medián': st.median(data),
  'szórás': st.stdev(data),
  'variancia': st.variance(data),
  'kvartilisek': st.quantiles(data),
}
statisztikák
```

## datetime – idő, dátum, ütemezés


```python
from datetime import datetime, date, time, timedelta

now = datetime.now()
print(now)
print(now.year, now.month, now.day)
print(now.hour, now.minute, now.second, now.microsecond)

# egyedi formázás
now.strftime("%Y-%m-%d %H:%M")
```

```python
end = now + timedelta(days=2, hours=5) # legyen ennyivel több
diff = end - now # vagy vissza különbséggé
diff.total_seconds() # összesen ennyi másodperc
```

Egyszerű ütemezés (futtasd másodpercenként)

```python
import time
interval = timedelta(seconds=1) # ennyi időközönként nézzük meg
next_time = datetime.now() + interval # legközelebb ekkor
timeout = datetime.now() + timedelta(seconds=5) # max ennyit várunk

while datetime.now() <= timeout:
    if datetime.now() >= next_time:
        print("Feladat fut:", datetime.now())
        next_time += interval # adunk hozzá időt
    time.sleep(0.2) # kicsit alszunk
```

## csv – mérési adatok beolvasása és kiírása
A legtöbb építőmérnöki adat (geodézia, labor, hidrológia) CSV formátumban érkezik.

Persze nem lenne különösebb kaland írni egy CSV beolvasót három sorban, de ez a csomag kezeli nekünk az apróságokat is (elválasztó karakter, fejléc, automatikus dict-konverzió, kódolás).

```python
# töltsünk le pár minta adatot csv-ben:
!wget 'https://gist.githubusercontent.com/goteguru/2e1efde943f9c963dcbb1fcae598b646/raw/0c37ffa0c7fc2ee77a539101ea7986f21a141fa8/sample_people.csv'
```

```python
import csv
with open("sample_people.csv") as f:
  rdr = csv.reader(f)
  összes = list(rdr) # mindet egy listába
  print(összes[:3]) # első három adat

```

```python
# adjuk most meg a kódolást is (bár az utf8 az alapértelmezés)
# és használjunk dict olvasót (a fejlécet használja indexeléshez!)
with open("sample_people.csv", newline="", encoding="utf8") as f:
  rdr = csv.DictReader(f)

  # soronként is feldolgozhatjuk
  for row in rdr:
      print(row["name"], row["city"], row["gender"])


```

```python
with open("sample_people.csv") as f:
  # vagy ha elfér a ramban...
  # mindet egy listába
  teljes  = list(csv.DictReader(f))

# és most kiírjuk magyar formátumban, pontosvesszővel, fejléccel:
fieldnames = ["név", "kor", "magasság"]
with open("kimenet.csv", "w") as f:
    wr = csv.DictWriter(f, fieldnames=fieldnames, delimiter=';')
    wr.writeheader()
    for r in teljes:
      wr.writerow({
          "név": r['name'],
          "magasság": r['height'], # felcseréltük
          "kor": r['age'] # de tudja a sorrendet
      })
```
