# Compound data structures

In addition to simple numeric and string data, Python provides several compound data structures.
These are essential because they are used constantly in real programs.

The most important built-in compound data structures are:

- **list** – ordered, mutable sequence
- **set** – unordered collection of unique elements
- **dict** – key–value mapping
- **tuple** – ordered, immutable sequence

All of these are *iterable*, meaning we can iterate over their elements one by one.
Because they store multiple values, we can often convert between them.

## Lists (`list`)

A list is an ordered, mutable data structure.
Order matters, elements can be changed, and duplicates are allowed.
Lists are indexed similarly to strings.

```python
# List literal using square brackets
values = [2, 8.4, "text"]

# Order matters when comparing lists
[2, 4, 1] == [2, 1, 4]
```

```python
# Lists are indexed starting from zero
numbers = [10, 20, 30, 50, 60]
numbers[0]
```

```python
# Negative index counts from the end
numbers[-1]
```

```python
# Slicing returns a list
numbers[0:2]
numbers[:]
numbers[-2:]
```

```python
# Reversed slice bounds result in an empty list
numbers[2:1]
```

```python
# Converting other iterables to a list
list("text")
list(("a", 5))
list({9, 8, 3, 1})
list(range(10, 20, 2))
```

```python
# List concatenation
[1, "Peter", 8] + ["Lajos", 42]
```

```python
# Lists are mutable
numbers = [10, 20, 30]
numbers[0] = 100
numbers
```

```python
# Appending an element
numbers.append("something")
numbers
```

```python
# Removing elements using del
del numbers[1:3]
numbers
```

# Tuple (N-tuple data)

A tuple is very similar to a list, but unlike a list it cannot be modified. It can contain different data types and can be of any length. However, once it is created, it stays that way. At most, we can create a new one from it.

In return, it uses less memory and is slightly faster. Because a tuple is immutable (it never changes), it is also `hashable`, so it can be used as a key (see below). Use it when this matters, but most importantly remember that values listed with commas (often inside parentheses) represent a tuple, because this form is used very frequently in libraries.

When to use it:
- If you explicitly want it not to change.
- If you want to use it as a key in a dict or as an element in a set.
- If you are storing a logically related "package of values" (e.g. coordinates).

Advantages:
- Faster and more memory-efficient than a list.
- Safe, because it cannot be modified.

Disadvantages:
- It cannot be edited later; a new one must be created.

```python
data_pair = ("Ákos", 22)
data_pair
```

```python
# order also matters in tuples:
("Peter", "Kinga") == ("Kinga", "Peter") # these are not equal
```

```python
# we can index them just like lists or strings:
data_pair[0] # first
data_pair[-1] # last
```

```python
# we can convert other things into a fixed-length tuple:
tuple("text") # from string to tuple
tuple([3,4,5]) # from list to tuple
```

We often see parentheses around tuples. Since parentheses are also used to define the order of operations, this can be confusing. For example, what is (-4)? Is it the number -4 put in parentheses just to make it clear that it is negative, or is it a single-element tuple?

From Python’s point of view, the essence of a tuple is not the parentheses (they are often not even required), but the comma. Something is only considered a tuple if there is a comma in it. If there is no comma, it is not a tuple. Parentheses are usually added only to avoid confusion with other constructs, such as listing function parameters.

```python
# this is a single-element tuple
(-5,) # we do not write anything after the comma; it is only there to make it a tuple
```

```python
# you could also write it like this; it is still a single-element tuple, just a bit more confusing
-5,
```

```python
# this, however, is the number -5 in parentheses. This is not a tuple.
(-5)
```

A tuple can store all kinds of data. Even data that itself is mutable. The tuple itself is still immutable!

```python
# A tuple can contain many different kinds of data:
(("another","tuple"), ["list",9], 99, "peter")
```

```python
# the fact that a tuple is immutable does not mean
# that the data stored inside it cannot be changed:

data = ([], [9]) # a pair consisting of an empty list and a single-element list
print(data)
data[0].append(100) # add an element to the first list
data[1][0] = 99 # overwrite the first (and only) element of the second list
print(data)
```

```python
# the tuple itself is what cannot be modified!
# remove the comment and you will get a nice big error!

# data[0] = [2,3,4] # overwrite the first list with another one
```

Tuples are often used to assign values to multiple variables at once. If there are multiple labels listed with commas on the left-hand side of the assignment, Python tries to "unpack" the compound structure on the right-hand side and assign its elements one by one.

```python
a, b, c = (1,2,3) # assign the three numbers at once!

b # so b gets the value 2
```

```python
# the essence of a tuple is the comma, not the parentheses; on the right-hand side
# of an assignment it is recognized even without parentheses, so you can safely omit them:
a, b, c = 1, 2, 3 # you can also write it this way
```

```python
# if Python cannot unpack it because the number of elements does not match, we get an error
a, b = 1, 2, 3 # cannot have more... <---- Error!
a, b, c = 1, 2 # and cannot have fewer either!  <---- Error!
```

## Set
Sets are comma-separated values enclosed in curly braces. Similar to the previous structures, they can contain values of any type (as long as they are "hashable").

The essence of a set is that it does not really care about order or multiplicity, only about whether something is a member or not. So if you add the same element multiple times, it will only be kept once (and the order is not preserved).

## Dictionaries (`dict`)

Lists and dictionaries are probably the most frequently used Python container data structures. A dictionary (or `dict` in Python) stores key → value pairs, meaning it is similar to a real dictionary, except that the key does not have to be a word – it can practically be anything (as long as it is `hashable`).

The important point is that a dictionary can contain each key exactly once, but the values associated with the keys may be the same. If the same key is specified multiple times, the later one overwrites the earlier one.

Its notation is similar to that of a set (comma-separated items inside curly braces), but here each item consists of two parts separated by a colon: a key and a value.

The purpose of a dictionary is that it can efficiently (quickly) find the value associated with a given key. It is well suited, for example, for storing configuration data or caching values that are expensive or slow to compute.

```python
# this is what a dictionary looks like:
dictionary = { "python": 100, "c": 100, "java": 68, "php": 30 }

# but it is often written across multiple lines for better readability:
dictionary = {
    "python": 100,
    "c":      100,  # <--- two values can be the same (100)
    "java":   68,
    "c":      99,   # <--- if the key is the same, only the later one counts
    "php":    30,   # <--- you may keep the trailing comma, it is not an error
}
dictionary
```

```python
# you index a dictionary using its keys:
dictionary["java"]
# and then it finds the associated value
```

```python
# and we can overwrite values just like in the case of lists:

dictionary["python"] = 200
dictionary
```

```python
# if we assign a value to a key that does not yet exist, it is not an error:
dictionary["fortran"] = 77 # now this exists as well...
dictionary
```

```python
# on the other hand, trying to access a non-existing key results in a KeyError!
# dictionary["french"] # <-- remove the comment to get an error.
```

Dictionaries contain key–value pairs, but if we only need one of them (only the keys or only the values), we can query those separately:

```python
dictionary.keys() # keys only
```

```python
dictionary.values() # values only
```

The *value* of a dictionary can be anything (for example a set or a list), but the key must always be a `hashable` value.

```python
weird_dict = {
    (4, 5): {9, 2}, # <-- key: tuple, value: set
    "Zsófia": 8,   # <-- key: str, value: int
    # [3]: 2.5,    # <-- this is not allowed because the key is a list
}
```

But what happens if we try to access a key that does not exist? Unfortunately, we get a `KeyError`. This is not always convenient, because it can easily happen that we do not know in advance whether a given key exists in the dictionary (for example, when the data comes from an external source, a database, or a network). What can we do in such cases?

```python
data = {"name": "Péter Nagy", "height": 177}

# this would raise an error because the "age" key is missing:
# data["age"]

# the get method helps us:
age = data.get("age") # if it does not exist, we simply get None instead of an error
print("age:", age) # print it (to see that it is None)

# if we do not like None, we can provide a default value ourselves!
city = data.get("city", "Budapest")
print("city:", city) # if there is no "city" key, it will be "Budapest"
```

If we are specifically interested in whether a given key exists (and not in its value), there is a dedicated operator for that: `in`!

```python
"name" in data # is there a "name" key in the data dictionary?
```

```python
"age" in data
```

# "Hashable" and immutability

If you are curious why there is all this fuss about `hashable` values, the explanation is actually very simple: something is called hashable if a "hash" can be created from it, that is, a constant value that never changes. Now, from a structure that changes all the time, such as a list or a set, it is quite difficult to create a constant value, because it may change at any moment. Dictionaries (their keys) and sets rely on the existence of such a constant value (the hash) in order to perform fast lookups. If there is no hash, they cannot do that. That is all there is to it.

This is why it is useful (among other reasons) that tuples and strings are immutable (cannot change), because then they can safely have a hash.

Since it is rather annoying that we cannot put sets inside sets (which would otherwise be quite logical), modern Python introduced an immutable version of a set that *is* `hashable`. This is the `frozenset`.

```python
a = frozenset((5, 5, 5, 7, 7, 7))
b = frozenset((5, 10))

{ a, b, a | b, a - b, a & b } # <- perfectly fine, a frozenset can be an element of a set
```

```python
# obviously, you cannot modify a frozenset
# a.add(9) # <-- this will not work...
```

When we get to classes, we will see that Python gives us a great deal of freedom: in fact, anything can be `hashable` if it has a `__hash__` method – that is, if we can define one.

# Looking ahead: Other types

These are Python’s built-in compound (container) types. If we need something more sophisticated (or more efficient), that is not a problem either – we can choose from countless libraries that provide additional data types.

The `collections` module contains many such structures:

- Counter: element counter (like a set, but it keeps track of counts)
- OrderedDict: like a dict, but order matters
- defaultdict: like a dict, but it automatically creates values for missing keys
- deque: like a list, useful when you mostly add/remove elements at the beginning or end
- namedtuple: like a tuple, but the values can be accessed by name instead of index

In the `array` package, we can find a kind of "list" that can only contain elements of the same type, but is more efficient (programmers usually call this an array). For the same purpose, we can also use `numpy`, which we will work with later!

Of course, there are also solutions that support more complex structures (for example in the `heapq` or `bisect` modules), but we will leave those to more advanced programmers :).
